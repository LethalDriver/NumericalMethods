```python
import numpy as np

def f1(x, v):
    return -x

def f2(x, v):
    return v

# Midpoint method implementation
def euler_method(x, y, dt):
    x1 = x + f2(x, y) * dt
    y1 = y + f1(x, y) * dt
    return x1, y1

def phase_portrait(x0, y0, dt=0.01, t_max=10):
    x, y = x0, y0
    xs, ys = [x], [y]
    for _ in range(int(t_max / dt)):
        x, y = euler_method(x, y, dt)
        xs.append(x)
        ys.append(y)
    return xs, ys

# Define initial conditions and parameters
x0, y0 = 1.0, 0.0  # Initial conditions

dt = 0.0001         # Time step
t_max = 2 * np.pi  # Maximum time

# Use the phase_portrait function to solve the system
xs, ys = phase_portrait(x0, y0, dt, t_max)

# Plot the results
import matplotlib.pyplot as plt

plt.plot(xs, ys)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase Portrait')
plt.show()
```
